```python
import pandas as pd

df = pd.read_csv('/content/ai_vs_human_dataset_medium.csv')

print("Первые 5 строк DataFrame:")
print(df.head())

print("\nПоследние 5 строк DataFrame:")
print(df.tail())

print("\nИнформация о DataFrame:")
df.info()
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
print("Типы данных после преобразования 'timestamp':")
df.info()
```

```python
print("Количество пропущенных значений до обработки:")
print(df.isnull().sum())
```

```python
df['notes'].fillna('Unknown', inplace=True)
print("\nКоличество пропущенных значений после обработки:")
print(df.isnull().sum())
```

```python
df['notes'] = df['notes'].fillna('Unknown')
print("\nКоличество пропущенных значений после обработки:")
print(df.isnull().sum())
```

```python
print(f"Количество дублирующихся строк до удаления: {df.duplicated().sum()}")
df.drop_duplicates(inplace=True)
print(f"Количество строк после удаления дубликатов: {len(df)}")
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

numeric_cols = ['length_chars', 'length_words', 'quality_score', 'sentiment', 'plagiarism_score']

print("Библиотеки импортированы и числовые столбцы определены.")
```

```python
outliers_count = {}
threshold = 3

for col in numeric_cols:
    z_scores = np.abs((df[col] - df[col].mean()) / df[col].std())
    outliers_count[col] = (z_scores > threshold).sum()

print("Количество обнаруженных выбросов (Z-score > 3):")
for col, count in outliers_count.items():
    print(f"Столбец '{col}': {count} выбросов")
```

```python
plt.figure(figsize=(15, 10))
for i, col in enumerate(numeric_cols):
    plt.subplot(2, 3, i + 1) # Настройте сетку подграфиков по мере необходимости
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot for {col}')
plt.tight_layout()
plt.show()
```

```python
print("\nОписательная статистика для числовых столбцов:")
print(df.describe())
```

```python
print("\nАнализ уникальных значений и их частот:")
for col in df.columns:
    unique_count = df[col].nunique()
    print(f"Столбец '{col}': {unique_count} уникальных значений")
    if df[col].dtype == 'object' and unique_count < 20:
        print(f"Частотное распределение для '{col}':\n{df[col].value_counts()}\n")
```

```python
print(
"\nМатрица корреляции для числовых столбцов:")
print(df.corr(numeric_only=True))
```

```python
plt.figure(figsize=(18, 12))

# 1. Гистограммы для 'length_chars' и 'quality_score'
plt.subplot(3, 3, 1)
sns.histplot(df['length_chars'], kde=True)
plt.title('Распределение длины (символов)')

plt.subplot(3, 3, 2)
sns.histplot(df['quality_score'], kde=True)
plt.title('Распределение оценки качества')
```

```python
df_sorted = df.sort_values(by='timestamp')

# 2. Линейный график для 'sentiment' по 'timestamp'
plt.subplot(3, 3, 3)
sns.lineplot(x=df_sorted['timestamp'], y=df_sorted['sentiment'])
plt.title('Тенденция сентимента со временем')
plt.xlabel('Временная метка')
plt.ylabel('Сентимент')
```

```python
# 3. Столбчатая диаграмма для подсчета 'label'
plt.subplot(3, 3, 4)
sns.countplot(x=df['label'])
plt.title('Количество меток (AI против Человека)')
plt.xlabel('Метка')
plt.ylabel('Количество')
```

```python
plt.subplot(3, 3, 5)
sns.barplot(x='label', y='quality_score', data=df)
plt.title('Средняя оценка качества по метке')
plt.xlabel('Метка')
plt.ylabel('Средняя оценка качества')
```

```python
plt.subplot(3, 3, 6)
sns.boxplot(x='label', y='quality_score', data=df)
plt.title('Распределение оценки качества по метке')
plt.xlabel('Метка')
plt.ylabel('Оценка качества')
```

```python
plt.subplot(3, 3, 7)
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Тепловая карта корреляционной матрицы')

plt.tight_layout()
plt.show()
```
